#Multigraficos (lineas, columnas)

```python
import matplotlib.pyplot as plt

# Crear la figura y la cuadrcula de subplots
figura, graficos = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))  # Aumentar tamao de la figura

# Grfico en la posicin (0,0)
graficos[0][0].plot(x, y, color="blue")
graficos[0][0].set_title("Grfico Azul")
graficos[0][0].set_xlabel("Eje X")
graficos[0][0].set_ylabel("Eje Y")

# Grfico en la posicin (0,1)
graficos[0][1].plot(x, y, color="red")
graficos[0][1].set_title("Grfico Rojo")
graficos[0][1].set_xlabel("Eje X")
graficos[0][1].set_ylabel("Eje Y")

# Grfico en la posicin (1,0)
graficos[1][0].plot(x, y, color="green")
graficos[1][0].set_title("Grfico Verde")
graficos[1][0].set_xlabel("Eje X")
graficos[1][0].set_ylabel("Eje Y")

# Grfico en la posicin (1,1)
graficos[1][1].plot(x, y, color="yellow")
graficos[1][1].set_title("Grfico Amarillo")
graficos[1][1].set_xlabel("Eje X")
graficos[1][1].set_ylabel("Eje Y")

# Ajustar el espacio entre los grficos
figura.tight_layout()

plt.show()
```

#Explicacin de los cambios:
#figsize=(10, 8) ajusta el tamao de la figura.
#set_title, set_xlabel, set_ylabel aaden ttulos y etiquetas a cada grfico.
#figura.tight_layout() ajusta el espacio entre subplots automticamente para evitar superposicin.

#Tamao Grafico

```python
import matplotlib.pyplot as plt

# Crear la figura y definir su tamao
figura = plt.figure(figsize=(8, 6))  # Establece el tamao de la figura en pulgadas (ancho=8, alto=6)

# Aadir un grfico en la figura usando coordenadas manuales para su posicin
# El formato [x, y, ancho, alto] est en porcentajes de la figura completa (0 a 1)
grafico1 = figura.add_axes([0.1, 0.1, 0.8, 0.8])  # Deja mrgenes alrededor del grfico principal

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear la lnea en el grfico con color y estilo personalizado
grafico1.plot(x, y, color="blue", marker="o", linestyle="-", label="Datos de ejemplo")  # Estilo de lnea y marcador

# Aadir ttulo y etiquetas de ejes para mejorar la presentacin
grafico1.set_title("Grfico con Tamao y Posicin Personalizada")  # Ttulo del grfico
grafico1.set_xlabel("Eje X - Ejemplo")  # Etiqueta del eje X
grafico1.set_ylabel("Eje Y - Ejemplo")  # Etiqueta del eje Y

# Activar cuadrcula para hacer ms fcil la lectura de valores
grafico1.grid(True, linestyle="--", alpha=0.6)  # Cuadrcula con estilo de lneas punteadas y semitransparente

# Mostrar la leyenda
grafico1.legend()

# Mostrar el grfico final
plt.show()
```

Explicacin de las mejoras:
Espacio alrededor del grfico (add_axes([0.1, 0.1, 0.8, 0.8])): Ajust el tamao para dejar un pequeo margen alrededor del grfico, asegurando que etiquetas y otros elementos no queden muy cerca del borde.
Personalizacin del estilo de la lnea: Aad marker="o" para resaltar cada punto y color="blue" para definir el color de la lnea, hacindolo ms visible.
Ttulo y etiquetas: Agregu un ttulo y etiquetas a los ejes (set_title, set_xlabel, set_ylabel) para mejorar la claridad del grfico.
Cuadrcula (grid): Activa una cuadrcula sutil que facilita la interpretacin de los valores en el grfico.
Leyenda: La legend() muestra una leyenda para identificar la lnea de datos, til cuando hay varias lneas o series de datos.

#Dos graficos en la misma Figura

```python
import matplotlib.pyplot as plt

# Crear la figura y definir el tamao para una mejor visualizacin
figura = plt.figure(figsize=(8, 6))

# Crear un conjunto de ejes en la figura con mrgenes alrededor del grfico
# Las coordenadas [0.1, 0.1, 0.8, 0.8] dejan espacio alrededor del grfico
graficos = figura.add_axes([0.1, 0.1, 0.8, 0.8])

# Datos de ejemplo
x = [1, 2, 3, 4, 5]

# Crear la primera lnea para x al cuadrado con una etiqueta descriptiva
graficos.plot(x, [i**2 for i in x], label="Al cuadrado", color="blue", marker="o", linestyle="-") 

# Crear la segunda lnea para x al cubo con una etiqueta descriptiva
graficos.plot(x, [i**3 for i in x], label="Al cubo", color="green", marker="s", linestyle="--")

# Configurar lmites de los ejes para evitar desfase y asegurar que todos los datos sean visibles
graficos.set_xlim([min(x), max(x)])  # Limitar el eje X al rango de valores de x
graficos.set_ylim([0, max(i**3 for i in x) * 1.1])  # Limitar el eje Y y dar un poco de espacio adicional arriba

# Aadir ttulo y etiquetas de ejes para mejorar la comprensin del grfico
graficos.set_title("Grficos de $x^2$ y $x^3$")  # Ttulo que describe el grfico
graficos.set_xlabel("Valores de X")  # Etiqueta para el eje X
graficos.set_ylabel("Valores de Y")  # Etiqueta para el eje Y

# Aadir una cuadrcula para facilitar la lectura
graficos.grid(True, linestyle="--", alpha=0.7)  # Estilo de lnea y transparencia en la cuadrcula

# Mostrar la leyenda en la mejor ubicacin (loc=0 para el mejor ajuste automtico)
graficos.legend(loc=0)

# Mostrar el grfico completo
plt.show()
```

#Color del Grafico

```python
import matplotlib.pyplot as plt

# Crear la figura principal y definir el tamao
figura = plt.figure(figsize=(8, 6))  # Tamao ajustado para mejor visualizacin

# Crear el grfico principal dentro de la figura, especificando su posicin y tamao
# Las coordenadas [0.1, 0.1, 0.8, 0.8] especifican la posicin (x, y) y el tamao (ancho, alto) como proporcin de la figura
grafico1 = figura.add_axes([0.1, 0.1, 0.8, 0.8])

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar los datos con un color verde degradado y estilo de lnea personalizado
grafico1.plot(x, y, color="seagreen", linestyle="--", marker="o", linewidth=2.5, label="Crecimiento cuadrtico")  
# `color` define el color de la lnea, `linestyle` el tipo de lnea, `marker` el marcador de cada punto
# `linewidth` ajusta el grosor de la lnea

# Aadir ttulo y etiquetas a los ejes para mayor claridad
grafico1.set_title("Crecimiento Cuadrtico con Color Personalizado")  # Ttulo descriptivo
grafico1.set_xlabel("Eje X - Tiempo")  # Etiqueta para el eje X
grafico1.set_ylabel("Eje Y - Valor")  # Etiqueta para el eje Y

# Agregar una cuadrcula para mejorar la lectura de valores
grafico1.grid(True, linestyle=":", alpha=0.7)  # Cuadrcula con estilo punteado y transparencia

# Aadir la leyenda en la mejor ubicacin para identificar la lnea de datos
grafico1.legend(loc="best")

# Mostrar el grfico completo
plt.show()
```

Explicacin de las mejoras:
Color (color="seagreen"): Us seagreen para un tono de verde ms sutil y agradable, pero puedes cambiarlo por cualquier otro nombre de color que Matplotlib acepte.

Estilo de lnea (linestyle="--"): El linestyle permite usar una lnea punteada, lo cual le da un efecto visual interesante. Puedes probar otros estilos como "-" para lnea continua o "-." para un estilo de lnea y punto.

Marcadores (marker="o"): Agregu un marcador circular ("o") en cada punto para destacar los datos individuales. Puedes cambiarlo a "s" para cuadrados o "^" para tringulos, segn prefieras.

Grosor de lnea (linewidth=2.5): Aument el grosor de la lnea para hacerla ms visible.

Ttulo y etiquetas de los ejes: Aad un ttulo claro y etiquetas a los ejes para hacer que el grfico sea ms descriptivo y fcil de entender.

Leyenda (legend): La leyenda describe la lnea del grfico y se ubica automticamente en la mejor posicin (loc="best").

```python
import matplotlib.pyplot as plt

# Crear la figura principal y definir el tamao para que sea fcil de visualizar
figura = plt.figure(figsize=(8, 6))  # Tamao de figura para evitar que la lnea se vea comprimida

# Crear el grfico principal ocupando todo el espacio en la figura
# [0, 0, 1, 1] hace que el grfico ocupe toda la figura sin mrgenes
grafico1 = figura.add_axes([0, 0, 1, 1])

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar los datos con personalizacin en la lnea:
grafico1.plot(x, y, 
              color="blue",        # Color de la lnea
              linewidth=10,         # Grosor de la lnea
              alpha=0.4,            # Transparencia de la lnea (1 es opaco, 0 es invisible)
              linestyle="-")        # Tipo de lnea: "-" para lnea continua

# Opcional: Agregar ttulo y etiquetas de ejes para proporcionar contexto al grfico
grafico1.set_title("Ejemplo de Personalizacin de Lnea en el Grfico")  # Ttulo descriptivo
grafico1.set_xlabel("Eje X - Valores")  # Etiqueta para el eje X
grafico1.set_ylabel("Eje Y - Resultados")  # Etiqueta para el eje Y

# Mostrar el grfico
plt.show()
```

#Marcadores

```python
figura = plt.figure()
grafico1 = figura.add_axes([0,0,1,1])
grafico1.plot(x,y,color="green", marker="o", markersize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Crear la figura principal y definir el tamao para una buena visualizacin de los marcadores
figura = plt.figure(figsize=(8, 6))  # Tamao de la figura

# Crear el grfico principal ocupando toda la figura
grafico1 = figura.add_axes([0, 0, 1, 1])  # Ocupa toda el rea de la figura

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar los datos con personalizacin enfocada en los marcadores
grafico1.plot(x, y, 
              color="green",            # Color de la lnea principal
              marker="o",               # Tipo de marcador (crculos en este caso)
              markersize=10,            # Tamao del marcador
              markerfacecolor="yellow",  # Color de relleno del marcador
              markeredgecolor="black",   # Color del borde del marcador
              markeredgewidth=1.5,       # Grosor del borde del marcador
              linestyle="-",             # Estilo de lnea: continuo
              linewidth=2,               # Grosor de la lnea principal
              alpha=0.8)                 # Transparencia de la lnea y marcadores

# Aadir ttulo y etiquetas de ejes para mejorar la claridad del grfico
grafico1.set_title("Grfico con Marcadores Personalizados")  # Ttulo del grfico
grafico1.set_xlabel("Eje X - Valores")  # Etiqueta para el eje X
grafico1.set_ylabel("Eje Y - Resultados")  # Etiqueta para el eje Y

# Mostrar el grfico completo
plt.show()
```

Explicacin de las personalizaciones de los marcadores:
Tipo de marcador (marker="o"): Aqu, "o" indica marcadores en forma de crculo, pero Matplotlib permite varios tipos, como "s" para cuadrados, "^" para tringulos, o "*". Cambiar el tipo de marcador puede ser til si necesitas distinguir entre diferentes conjuntos de datos en el mismo grfico.

Tamao del marcador (markersize=10): Establece el tamao del marcador. Un tamao ms grande (10) hace que los puntos de datos individuales sean ms visibles.

Color de relleno (markerfacecolor="yellow"): El color de relleno en los marcadores se establece en amarillo, haciendo que destaquen sobre la lnea verde. Esto es til si quieres que los marcadores se diferencien visualmente de la lnea.

Color y grosor del borde (markeredgecolor="black", markeredgewidth=1.5): El borde negro del marcador ayuda a definir sus bordes y mejorar la visibilidad en grficos con lneas y fondos complejos. El grosor de 1.5 agrega un contorno visible, ayudando a distinguir los puntos.

Transparencia (alpha=0.8): Ajusta la opacidad tanto de los marcadores como de la lnea principal. Un valor de 0.8 es til para suavizar la visualizacin sin perder demasiada visibilidad.

Ejercicio
crea una variable "x" que tenga 10 valores equidistantes entre los valores 0 y 20 
crar una variable "y" que tenmga los valores de la funcion "x al cuadrado +5"
Crear un grafio de color rojo, con ancho de 5 y estilo de puntos "...."

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear la variable `x` con 10 valores equidistantes entre 0 y 20
x = np.linspace(0, 20, 10)

# Crear la variable `y` como la funcin `y = x^2 + 5`
y = x**2 + 5

# Crear el grfico con los requisitos especificados
plt.plot(x, y, color="red", linewidth=5, linestyle=":")  # Estilo de puntos

# Aadir ttulo y etiquetas a los ejes para mayor claridad
plt.title("Grfico de $y = x^2 + 5$")  # Ttulo del grfico
plt.xlabel("Valores de X")  # Etiqueta para el eje X
plt.ylabel("Valores de Y")  # Etiqueta para el eje Y

# Mostrar el grfico
plt.show()
```

Explicacin del cdigo:
x = np.linspace(0, 20, 10): Utiliza linspace de numpy para crear 10 valores equidistantes entre 0 y 20.
Definicin de y: La funcin 
=2+5 se calcula usando la potencia de x ms 5.
Estilo del grfico: color="red", linewidth=5, y linestyle=":" producen una lnea roja, gruesa y con un estilo de puntos.
Ttulo y etiquetas de ejes: Estas adiciones ayudan a interpretar el grfico y entender el contexto de la funcin graficada.
