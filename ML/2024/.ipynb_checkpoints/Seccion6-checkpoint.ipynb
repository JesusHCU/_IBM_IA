```python
import matplotlib.pyplot as plt

# Crear la figura principal
figura = plt.figure(figsize=(8, 6))  # Configuramos el tamaño de la figura

# Crear el gráfico principal que ocupa la mayor parte de la figura
grafico1 = figura.add_axes([0.1, 0.1, 0.8, 0.8])  # [x_inicial, y_inicial, ancho, alto] en porcentaje
grafico1.plot(x, y, color="red", label="Datos principales")  # Dibujar línea roja
grafico1.set_title("Gráfico Principal")  # Título para el gráfico principal
grafico1.set_xlabel("Eje X")  # Etiqueta del eje X
grafico1.set_ylabel("Eje Y")  # Etiqueta del eje Y
grafico1.legend()  # Mostrar leyenda

# Crear el gráfico secundario como inserción en la parte superior derecha
grafico2 = figura.add_axes([0.2, 0.45, 0.3, 0.3])  # Ajustado para un buen encaje en la figura
grafico2.plot(x, y, color="green", label="Subgráfico")  # Dibujar línea verde
grafico2.set_title("Subgráfico Insertado")  # Título para el gráfico secundario
grafico2.set_xlabel("Eje X (Zoom)")  # Etiqueta del eje X en subgráfico
grafico2.set_ylabel("Eje Y (Zoom)")  # Etiqueta del eje Y en subgráfico
grafico2.legend()  # Mostrar leyenda

# Mostrar la figura completa con ambos gráficos
plt.show()



#Explicación de las mejoras y comentarios:
#Tamaño de la figura: Definimos el tamaño de figura usando figsize, lo que asegura que los gráficos tengan una buena escala y se vean bien.
#Ajuste de los ejes: Ajustamos las dimensiones de los gráficos para que se adapten de manera armoniosa en la figura.
#Títulos y etiquetas: Añadimos títulos (set_title) y etiquetas (set_xlabel, set_ylabel) para hacer más claro el contenido de cada gráfico.
#Leyendas: Se añadieron leyendas con legend() para identificar cada línea de datos.
```

#Multigraficos (lineas, columnas)

```python
import matplotlib.pyplot as plt

# Crear la figura y la cuadrícula de subplots
figura, graficos = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))  # Aumentar tamaño de la figura

# Gráfico en la posición (0,0)
graficos[0][0].plot(x, y, color="blue")
graficos[0][0].set_title("Gráfico Azul")
graficos[0][0].set_xlabel("Eje X")
graficos[0][0].set_ylabel("Eje Y")

# Gráfico en la posición (0,1)
graficos[0][1].plot(x, y, color="red")
graficos[0][1].set_title("Gráfico Rojo")
graficos[0][1].set_xlabel("Eje X")
graficos[0][1].set_ylabel("Eje Y")

# Gráfico en la posición (1,0)
graficos[1][0].plot(x, y, color="green")
graficos[1][0].set_title("Gráfico Verde")
graficos[1][0].set_xlabel("Eje X")
graficos[1][0].set_ylabel("Eje Y")

# Gráfico en la posición (1,1)
graficos[1][1].plot(x, y, color="yellow")
graficos[1][1].set_title("Gráfico Amarillo")
graficos[1][1].set_xlabel("Eje X")
graficos[1][1].set_ylabel("Eje Y")

# Ajustar el espacio entre los gráficos
figura.tight_layout()

plt.show()
```

#Explicación de los cambios:
#figsize=(10, 8) ajusta el tamaño de la figura.
#set_title, set_xlabel, set_ylabel añaden títulos y etiquetas a cada gráfico.
#figura.tight_layout() ajusta el espacio entre subplots automáticamente para evitar superposición.

#Tamaño Grafico

```python
import matplotlib.pyplot as plt

# Crear la figura y definir su tamaño
figura = plt.figure(figsize=(8, 6))  # Establece el tamaño de la figura en pulgadas (ancho=8, alto=6)

# Añadir un gráfico en la figura usando coordenadas manuales para su posición
# El formato [x, y, ancho, alto] está en porcentajes de la figura completa (0 a 1)
grafico1 = figura.add_axes([0.1, 0.1, 0.8, 0.8])  # Deja márgenes alrededor del gráfico principal

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Crear la línea en el gráfico con color y estilo personalizado
grafico1.plot(x, y, color="blue", marker="o", linestyle="-", label="Datos de ejemplo")  # Estilo de línea y marcador

# Añadir título y etiquetas de ejes para mejorar la presentación
grafico1.set_title("Gráfico con Tamaño y Posición Personalizada")  # Título del gráfico
grafico1.set_xlabel("Eje X - Ejemplo")  # Etiqueta del eje X
grafico1.set_ylabel("Eje Y - Ejemplo")  # Etiqueta del eje Y

# Activar cuadrícula para hacer más fácil la lectura de valores
grafico1.grid(True, linestyle="--", alpha=0.6)  # Cuadrícula con estilo de líneas punteadas y semitransparente

# Mostrar la leyenda
grafico1.legend()

# Mostrar el gráfico final
plt.show()
```

Explicación de las mejoras:
Espacio alrededor del gráfico (add_axes([0.1, 0.1, 0.8, 0.8])): Ajusté el tamaño para dejar un pequeño margen alrededor del gráfico, asegurando que etiquetas y otros elementos no queden muy cerca del borde.
Personalización del estilo de la línea: Añadí marker="o" para resaltar cada punto y color="blue" para definir el color de la línea, haciéndolo más visible.
Título y etiquetas: Agregué un título y etiquetas a los ejes (set_title, set_xlabel, set_ylabel) para mejorar la claridad del gráfico.
Cuadrícula (grid): Activa una cuadrícula sutil que facilita la interpretación de los valores en el gráfico.
Leyenda: La legend() muestra una leyenda para identificar la línea de datos, útil cuando hay varias líneas o series de datos.

#Dos graficos en la misma Figura

```python
import matplotlib.pyplot as plt

# Crear la figura y definir el tamaño para una mejor visualización
figura = plt.figure(figsize=(8, 6))

# Crear un conjunto de ejes en la figura con márgenes alrededor del gráfico
# Las coordenadas [0.1, 0.1, 0.8, 0.8] dejan espacio alrededor del gráfico
graficos = figura.add_axes([0.1, 0.1, 0.8, 0.8])

# Datos de ejemplo
x = [1, 2, 3, 4, 5]

# Crear la primera línea para x al cuadrado con una etiqueta descriptiva
graficos.plot(x, [i**2 for i in x], label="Al cuadrado", color="blue", marker="o", linestyle="-") 

# Crear la segunda línea para x al cubo con una etiqueta descriptiva
graficos.plot(x, [i**3 for i in x], label="Al cubo", color="green", marker="s", linestyle="--")

# Configurar límites de los ejes para evitar desfase y asegurar que todos los datos sean visibles
graficos.set_xlim([min(x), max(x)])  # Limitar el eje X al rango de valores de x
graficos.set_ylim([0, max(i**3 for i in x) * 1.1])  # Limitar el eje Y y dar un poco de espacio adicional arriba

# Añadir título y etiquetas de ejes para mejorar la comprensión del gráfico
graficos.set_title("Gráficos de $x^2$ y $x^3$")  # Título que describe el gráfico
graficos.set_xlabel("Valores de X")  # Etiqueta para el eje X
graficos.set_ylabel("Valores de Y")  # Etiqueta para el eje Y

# Añadir una cuadrícula para facilitar la lectura
graficos.grid(True, linestyle="--", alpha=0.7)  # Estilo de línea y transparencia en la cuadrícula

# Mostrar la leyenda en la mejor ubicación (loc=0 para el mejor ajuste automático)
graficos.legend(loc=0)

# Mostrar el gráfico completo
plt.show()
```

#Color del Grafico

```python
import matplotlib.pyplot as plt

# Crear la figura principal y definir el tamaño
figura = plt.figure(figsize=(8, 6))  # Tamaño ajustado para mejor visualización

# Crear el gráfico principal dentro de la figura, especificando su posición y tamaño
# Las coordenadas [0.1, 0.1, 0.8, 0.8] especifican la posición (x, y) y el tamaño (ancho, alto) como proporción de la figura
grafico1 = figura.add_axes([0.1, 0.1, 0.8, 0.8])

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar los datos con un color verde degradado y estilo de línea personalizado
grafico1.plot(x, y, color="seagreen", linestyle="--", marker="o", linewidth=2.5, label="Crecimiento cuadrático")  
# `color` define el color de la línea, `linestyle` el tipo de línea, `marker` el marcador de cada punto
# `linewidth` ajusta el grosor de la línea

# Añadir título y etiquetas a los ejes para mayor claridad
grafico1.set_title("Crecimiento Cuadrático con Color Personalizado")  # Título descriptivo
grafico1.set_xlabel("Eje X - Tiempo")  # Etiqueta para el eje X
grafico1.set_ylabel("Eje Y - Valor")  # Etiqueta para el eje Y

# Agregar una cuadrícula para mejorar la lectura de valores
grafico1.grid(True, linestyle=":", alpha=0.7)  # Cuadrícula con estilo punteado y transparencia

# Añadir la leyenda en la mejor ubicación para identificar la línea de datos
grafico1.legend(loc="best")

# Mostrar el gráfico completo
plt.show()
```

Explicación de las mejoras:
Color (color="seagreen"): Usé seagreen para un tono de verde más sutil y agradable, pero puedes cambiarlo por cualquier otro nombre de color que Matplotlib acepte.

Estilo de línea (linestyle="--"): El linestyle permite usar una línea punteada, lo cual le da un efecto visual interesante. Puedes probar otros estilos como "-" para línea continua o "-." para un estilo de línea y punto.

Marcadores (marker="o"): Agregué un marcador circular ("o") en cada punto para destacar los datos individuales. Puedes cambiarlo a "s" para cuadrados o "^" para triángulos, según prefieras.

Grosor de línea (linewidth=2.5): Aumenté el grosor de la línea para hacerla más visible.

Título y etiquetas de los ejes: Añadí un título claro y etiquetas a los ejes para hacer que el gráfico sea más descriptivo y fácil de entender.

Leyenda (legend): La leyenda describe la línea del gráfico y se ubica automáticamente en la mejor posición (loc="best").

```python
import matplotlib.pyplot as plt

# Crear la figura principal y definir el tamaño para que sea fácil de visualizar
figura = plt.figure(figsize=(8, 6))  # Tamaño de figura para evitar que la línea se vea comprimida

# Crear el gráfico principal ocupando todo el espacio en la figura
# [0, 0, 1, 1] hace que el gráfico ocupe toda la figura sin márgenes
grafico1 = figura.add_axes([0, 0, 1, 1])

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar los datos con personalización en la línea:
grafico1.plot(x, y, 
              color="blue",        # Color de la línea
              linewidth=10,         # Grosor de la línea
              alpha=0.4,            # Transparencia de la línea (1 es opaco, 0 es invisible)
              linestyle="-")        # Tipo de línea: "-" para línea continua

# Opcional: Agregar título y etiquetas de ejes para proporcionar contexto al gráfico
grafico1.set_title("Ejemplo de Personalización de Línea en el Gráfico")  # Título descriptivo
grafico1.set_xlabel("Eje X - Valores")  # Etiqueta para el eje X
grafico1.set_ylabel("Eje Y - Resultados")  # Etiqueta para el eje Y

# Mostrar el gráfico
plt.show()
```

#Marcadores

```python
figura = plt.figure()
grafico1 = figura.add_axes([0,0,1,1])
grafico1.plot(x,y,color="green", marker="o", markersize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Crear la figura principal y definir el tamaño para una buena visualización de los marcadores
figura = plt.figure(figsize=(8, 6))  # Tamaño de la figura

# Crear el gráfico principal ocupando toda la figura
grafico1 = figura.add_axes([0, 0, 1, 1])  # Ocupa toda el área de la figura

# Datos de ejemplo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Graficar los datos con personalización enfocada en los marcadores
grafico1.plot(x, y, 
              color="green",            # Color de la línea principal
              marker="o",               # Tipo de marcador (círculos en este caso)
              markersize=10,            # Tamaño del marcador
              markerfacecolor="yellow",  # Color de relleno del marcador
              markeredgecolor="black",   # Color del borde del marcador
              markeredgewidth=1.5,       # Grosor del borde del marcador
              linestyle="-",             # Estilo de línea: continuo
              linewidth=2,               # Grosor de la línea principal
              alpha=0.8)                 # Transparencia de la línea y marcadores

# Añadir título y etiquetas de ejes para mejorar la claridad del gráfico
grafico1.set_title("Gráfico con Marcadores Personalizados")  # Título del gráfico
grafico1.set_xlabel("Eje X - Valores")  # Etiqueta para el eje X
grafico1.set_ylabel("Eje Y - Resultados")  # Etiqueta para el eje Y

# Mostrar el gráfico completo
plt.show()
```

Explicación de las personalizaciones de los marcadores:
Tipo de marcador (marker="o"): Aquí, "o" indica marcadores en forma de círculo, pero Matplotlib permite varios tipos, como "s" para cuadrados, "^" para triángulos, o "*". Cambiar el tipo de marcador puede ser útil si necesitas distinguir entre diferentes conjuntos de datos en el mismo gráfico.

Tamaño del marcador (markersize=10): Establece el tamaño del marcador. Un tamaño más grande (10) hace que los puntos de datos individuales sean más visibles.

Color de relleno (markerfacecolor="yellow"): El color de relleno en los marcadores se establece en amarillo, haciendo que destaquen sobre la línea verde. Esto es útil si quieres que los marcadores se diferencien visualmente de la línea.

Color y grosor del borde (markeredgecolor="black", markeredgewidth=1.5): El borde negro del marcador ayuda a definir sus bordes y mejorar la visibilidad en gráficos con líneas y fondos complejos. El grosor de 1.5 agrega un contorno visible, ayudando a distinguir los puntos.

Transparencia (alpha=0.8): Ajusta la opacidad tanto de los marcadores como de la línea principal. Un valor de 0.8 es útil para suavizar la visualización sin perder demasiada visibilidad.

Ejercicio
crea una variable "x" que tenga 10 valores equidistantes entre los valores 0 y 20 
crar una variable "y" que tenmga los valores de la funcion "x al cuadrado +5"
Crear un grafio de color rojo, con ancho de 5 y estilo de puntos "...."

```python
import numpy as np
import matplotlib.pyplot as plt

# Crear la variable `x` con 10 valores equidistantes entre 0 y 20
x = np.linspace(0, 20, 10)

# Crear la variable `y` como la función `y = x^2 + 5`
y = x**2 + 5

# Crear el gráfico con los requisitos especificados
plt.plot(x, y, color="red", linewidth=5, linestyle=":")  # Estilo de puntos

# Añadir título y etiquetas a los ejes para mayor claridad
plt.title("Gráfico de $y = x^2 + 5$")  # Título del gráfico
plt.xlabel("Valores de X")  # Etiqueta para el eje X
plt.ylabel("Valores de Y")  # Etiqueta para el eje Y

# Mostrar el gráfico
plt.show()
```

Explicación del código:
x = np.linspace(0, 20, 10): Utiliza linspace de numpy para crear 10 valores equidistantes entre 0 y 20.
Definición de y: La función 
𝑦=𝑥2+5 se calcula usando la potencia de x más 5.
Estilo del gráfico: color="red", linewidth=5, y linestyle=":" producen una línea roja, gruesa y con un estilo de puntos.
Título y etiquetas de ejes: Estas adiciones ayudan a interpretar el gráfico y entender el contexto de la función graficada.
