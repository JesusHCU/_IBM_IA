```python
#Ejemplo como lo indicado en clase

# Importamos las bibliotecas necesarias
import pandas as pd  # Para manejar datos en formato de DataFrame
import numpy as np  # Para operaciones numéricas
import seaborn as sns  # Para visualización de datos
import matplotlib.pyplot as plt  # Para graficar
from sklearn.datasets import make_blobs  # Para generar datos sintéticos
from sklearn.cluster import KMeans  # Para el algoritmo K-means

# Generación de datos sintéticos
# 'make_blobs' se utiliza para crear un conjunto de datos con un número específico de muestras, características y centros
n_samples = 200  # Número total de muestras
n_features = 2  # Número de características (en este caso, 2D)
centers = 4  # Número de centros/clústeres que queremos
datos = make_blobs(n_samples=n_samples, n_features=n_features, centers=centers, random_state=42)

# Visualizamos los datos generados sin aplicar el algoritmo K-means
plt.scatter(datos[0][:, 0], datos[0][:, 1], s=50)  # Graficamos los puntos
plt.title("Datos Generados")
plt.xlabel("Característica 1")
plt.ylabel("Característica 2")
plt.show()

# Creación y entrenamiento del modelo K-means
modelo = KMeans(n_clusters=centers, random_state=42)  # Inicializamos el modelo con el número de clústeres deseado
modelo.fit(datos[0])  # Ajustamos el modelo a los datos generados

# Obtención de los centroides de los clústeres
centroides = modelo.cluster_centers_  # Coordenadas de los centroides
etiquetas = modelo.labels_  # Etiquetas de clúster asignadas a cada muestra

# Visualización de los resultados del clustering
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))  # Creamos un gráfico con 2 subgráficas

# Gráfico 1: Resultados del K-means
ax1.scatter(datos[0][:, 0], datos[0][:, 1], c=etiquetas, s=50, cmap='viridis')  # Coloreamos los puntos según las etiquetas
ax1.scatter(centroides[:, 0], centroides[:, 1], c='red', s=200, alpha=0.75, marker='X')  # Agregamos los centroides
ax1.set_title("Resultados del Algoritmo K-means")
ax1.set_xlabel("Característica 1")
ax1.set_ylabel("Característica 2")

# Gráfico 2: Datos originales con etiquetas reales
ax2.scatter(datos[0][:, 0], datos[0][:, 1], c=datos[1], s=50, cmap='viridis')  # Coloreamos los puntos según las etiquetas reales
ax2.set_title("Datos Originales")
ax2.set_xlabel("Característica 1")
ax2.set_ylabel("Característica 2")

plt.show()  # Mostramos los gráficos
```

Usamos make_blobs para generar un conjunto de datos sintético. Este método es útil para simular datos donde sabemos la estructura (número de clústeres) que queremos estudiar.

Inicializamos el modelo KMeans con el número deseado de clústeres.
Ajustamos el modelo a los datos usando fit, lo que permite al modelo encontrar los clústeres en los datos.

En el gráfico de resultados de K-means, también mostramos los centroides con un símbolo de "X" rojo para indicar las posiciones centrales de cada clúster.
Notas sobre el Algoritmo K-means:
K-means es un algoritmo de clustering que busca agrupar n observaciones en k grupos en los que cada observación pertenece al clúster con la media más cercana (centroide).
El número de clústeres (k) debe ser especificado antes de ejecutar el algoritmo.
Es sensible a la inicialización y puede converger a soluciones locales, por lo que es común ejecutar el algoritmo varias veces con diferentes inicializaciones.

```python
# Importamos las bibliotecas necesarias
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# 1. Preparación de los datos
# Cargamos los datos. Asegúrate de que tienes un archivo CSV o usa el conjunto de datos Iris
data = pd.read_csv('iris.csv')  # Reemplaza 'iris.csv' con tu archivo de datos

# Supongamos que las columnas de características son "SepalLengthCm", "SepalWidthCm", "PetalLengthCm" y "PetalWidthCm".
# Si tienes una columna de etiquetas (por ejemplo, 'Species'), elimínala para trabajar solo con datos no etiquetados.
# Esto permite que el modelo K-means haga el agrupamiento sin supervisión.
X = data.drop("Species", axis=1)  # Mantiene solo las características numéricas

# 2. Escalado de los datos
# Escalamos las características para que estén en la misma escala, lo cual es esencial en K-means ya que usa la distancia Euclidiana.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Definir el número de clusters (k)
# Aquí elegimos el valor de 'k' (cantidad de clusters). Vamos a usar el método del codo para decidir el valor óptimo de k.

# Método del codo
inercia = []  # Lista para guardar la inercia (suma de las distancias al centroide más cercano para cada punto)
k_values = range(1, 11)  # Probamos k desde 1 hasta 10

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inercia.append(kmeans.inertia_)  # Guardamos la inercia para cada valor de k

# Graficamos el método del codo para observar dónde la inercia deja de disminuir significativamente
plt.figure(figsize=(8, 5))
plt.plot(k_values, inercia, 'bo-', markersize=8)
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo para elegir k')
plt.show()

# Generalmente, el punto donde se forma el "codo" (una caída brusca en la gráfica) sugiere el valor ideal de k.
# Para este ejemplo, supongamos que k = 3 se considera óptimo después de observar el gráfico.

# 4. Aplicar K-means con el k óptimo
k_optimo = 3
kmeans = KMeans(n_clusters=k_optimo, random_state=42)
kmeans.fit(X_scaled)

# Predicción de clusters para los datos
clusters = kmeans.predict(X_scaled)

# Añadimos los resultados del cluster al DataFrame original para analizar el agrupamiento
data['Cluster'] = clusters

# 5. Visualización de los clusters (opcional, en caso de tener 2 o 3 dimensiones)
# Aquí, usaremos un gráfico de dispersión para ver la distribución de los clusters
# Nota: Esto solo es útil si reducimos los datos a 2D con PCA o si nuestros datos tienen solo 2 características

plt.figure(figsize=(8, 5))
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='x', label='Centroides')
plt.xlabel('Primera característica (escalada)')
plt.ylabel('Segunda característica (escalada)')
plt.title('Agrupamiento de datos con K-means')
plt.legend()
plt.show()

# 6. Interpretación de los resultados
# Ahora el DataFrame 'data' tiene una nueva columna 'Cluster' que asigna cada punto a uno de los clusters obtenidos.
print(data.head())

# 7. Opciones y variaciones de K-means
# Puedes ajustar algunos parámetros del modelo, como:
# - `n_clusters`: Cambia el número de clusters según los resultados del método del codo.
# - `init`: Método para inicializar los centroides. Los valores comunes son 'k-means++' (mejora la eficiencia) y 'random' (más rápido).
# - `max_iter`: Número máximo de iteraciones para evitar bucles infinitos si no se alcanza la convergencia.
# - `n_init`: Número de veces que el algoritmo se ejecutará con diferentes inicializaciones de centroides (mejora la estabilidad de resultados).

# Variación: K-means MiniBatch para grandes conjuntos de datos
# from sklearn.cluster import MiniBatchKMeans
# mini_kmeans = MiniBatchKMeans(n_clusters=k_optimo, batch_size=100, random_state=42)
# mini_kmeans.fit(X_scaled)
```

```python
#Mejora
# Importamos las bibliotecas necesarias
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
import os

# Ajuste de entorno para evitar la advertencia de fugas de memoria en Windows con MKL
# Establecemos el número de hilos de OpenMP a 1, lo cual es una solución recomendada en Windows
os.environ["OMP_NUM_THREADS"] = "1"

# Función de clustering con K-means o MiniBatchKMeans
def kmeans_clustering(datos, n_clusters=3, use_mini_batch=False):
    """
    Esta función aplica el algoritmo K-means para agrupar datos, con opción a usar MiniBatchKMeans
    para mayor eficiencia en conjuntos de datos grandes.
    
    Parámetros:
    - datos (DataFrame): Conjunto de datos con las características a agrupar.
    - n_clusters (int): Número de clústeres para K-means.
    - use_mini_batch (bool): Si es True, usa MiniBatchKMeans en lugar de KMeans.
    
    Retorna:
    - modelo_kmeans: Modelo K-means o MiniBatchKMeans entrenado.
    - etiquetas: Etiquetas de clúster para cada muestra en el conjunto de datos.
    - centroides: Coordenadas de los centroides de los clústeres.
    """
    # Selección de características (quitamos cualquier columna de etiquetas si está presente)
    X = datos.drop("Species", axis=1) if "Species" in datos.columns else datos

    # Escalado de las características
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Selección del modelo: KMeans o MiniBatchKMeans según el parámetro `use_mini_batch`
    if use_mini_batch:
        # MiniBatchKMeans es más eficiente en memoria para grandes conjuntos de datos
        modelo_kmeans = MiniBatchKMeans(n_clusters=n_clusters, random_state=42, batch_size=100)
    else:
        # Modelo estándar de KMeans
        modelo_kmeans = KMeans(n_clusters=n_clusters, random_state=42)

    # Entrenamiento del modelo en los datos escalados
    modelo_kmeans.fit(X_scaled)

    # Obtención de etiquetas y centroides
    etiquetas = modelo_kmeans.labels_
    centroides = modelo_kmeans.cluster_centers_

    return modelo_kmeans, etiquetas, centroides

# Carga de datos y aplicación de la función
data = pd.read_csv('iris.csv')

# Ejecutamos KMeans con MiniBatchKMeans para evitar problemas de memoria en Windows
modelo_kmeans, etiquetas, centroides = kmeans_clustering(data, n_clusters=3, use_mini_batch=True)

# Mostrar los resultados de agrupamiento
print("Etiquetas de Clúster para cada muestra:")
print(etiquetas)

print("Centroides de los Clústeres:")
print(centroides)
```

Explicación de Modificaciones:
Variable de Entorno OMP_NUM_THREADS: Al inicio del código, se establece OMP_NUM_THREADS=1 para evitar problemas de memoria en Windows.

Opción para MiniBatchKMeans: Se ha agregado un parámetro use_mini_batch. Si se pasa como True, la función usa MiniBatchKMeans, que es más eficiente para conjuntos grandes y puede prevenir problemas de memoria.

Ejecución con MiniBatchKMeans: En la llamada de ejemplo, use_mini_batch=True se especifica para evitar las advertencias en Windows.º

```python
# Importamos las bibliotecas necesarias
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.preprocessing import StandardScaler
import os

# Ajuste de entorno para evitar la advertencia de fugas de memoria en Windows con MKL
os.environ["OMP_NUM_THREADS"] = "1"  # Establecemos los hilos a 1

# Función de clustering con K-means o MiniBatchKMeans
def kmeans_clustering(datos, n_clusters=3, use_mini_batch=False):
    """
    Esta función aplica el algoritmo K-means para agrupar datos, con opción a usar MiniBatchKMeans
    para mayor eficiencia en conjuntos de datos grandes.
    
    Parámetros:
    - datos (DataFrame): Conjunto de datos con las características a agrupar.
    - n_clusters (int): Número de clústeres para K-means.
    - use_mini_batch (bool): Si es True, usa MiniBatchKMeans en lugar de KMeans.
    
    Retorna:
    - modelo_kmeans: Modelo K-means o MiniBatchKMeans entrenado.
    - etiquetas: Etiquetas de clúster para cada muestra en el conjunto de datos.
    - centroides: Coordenadas de los centroides de los clústeres.
    """
    # Selección de características (quitamos cualquier columna de etiquetas si está presente)
    X = datos.drop("Species", axis=1) if "Species" in datos.columns else datos

    # Escalado de las características
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Selección del modelo: KMeans o MiniBatchKMeans según el parámetro `use_mini_batch`
    if use_mini_batch:
        # MiniBatchKMeans con batch_size>=3072 para evitar advertencia de fuga de memoria en Windows
        modelo_kmeans = MiniBatchKMeans(n_clusters=n_clusters, random_state=42, batch_size=3072)
    else:
        # Modelo estándar de KMeans
        modelo_kmeans = KMeans(n_clusters=n_clusters, random_state=42)

    # Entrenamiento del modelo en los datos escalados
    modelo_kmeans.fit(X_scaled)

    # Obtención de etiquetas y centroides
    etiquetas = modelo_kmeans.labels_
    centroides = modelo_kmeans.cluster_centers_

    return modelo_kmeans, etiquetas, centroides

# Carga de datos y aplicación de la función
data = pd.read_csv('iris.csv')

# Ejecutamos KMeans con MiniBatchKMeans para evitar problemas de memoria en Windows
modelo_kmeans, etiquetas, centroides = kmeans_clustering(data, n_clusters=3, use_mini_batch=True)

# Mostrar los resultados de agrupamiento
print("Etiquetas de Clúster para cada muestra:")
print(etiquetas)

print("Centroides de los Clústeres:")
print(centroides)
```

Opción 2: Alternativa sin MiniBatchKMeans
Si el problema persiste incluso con batch_size y la variable de entorno, podrías considerar usar el modelo estándar KMeans, que generalmente es más estable, a menos que estés trabajando con un conjunto de datos extremadamente grande. Aquí te muestro cómo implementarlo:

```python
# Importamos las bibliotecas necesarias
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

def kmeans_clustering(datos, n_clusters=3):
    """
    Función que aplica el algoritmo KMeans para agrupar datos.

    Parámetros:
    - datos (DataFrame): Conjunto de datos con las características a agrupar.
    - n_clusters (int): Número de clústeres para KMeans.

    Retorna:
    - modelo_kmeans: Modelo KMeans entrenado.
    - etiquetas: Etiquetas de clúster para cada muestra en el conjunto de datos.
    - centroides: Coordenadas de los centroides de los clústeres.
    """
    # Selección de características (quitamos cualquier columna de etiquetas si está presente)
    X = datos.drop("Species", axis=1) if "Species" in datos.columns else datos

    # Escalado de las características
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Inicialización y entrenamiento del modelo KMeans
    modelo_kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    modelo_kmeans.fit(X_scaled)

    # Obtención de etiquetas y centroides
    etiquetas = modelo_kmeans.labels_
    centroides = modelo_kmeans.cluster_centers_

    return modelo_kmeans, etiquetas, centroides

# Carga de datos y aplicación de la función
data = pd.read_csv('iris.csv')

# Ejecutamos el clustering usando KMeans estándar
modelo_kmeans, etiquetas, centroides = kmeans_clustering(data, n_clusters=3)

# Mostrar los resultados de agrupamiento
print("Etiquetas de Clúster para cada muestra:")
print(etiquetas)

print("Centroides de los Clústeres:")
print(centroides)
```
