```python
# Filtra el DataFrame por Género
#peliculas2 = peliculas[(peliculas.Genero == 'accion') | (peliculas.Genero == 'aventura') | (peliculas.Genero == 'animacion') | (peliculas.Genero == 'comedia') | (peliculas.Genero == 'drama')]
filtro_generos=['accion', 'aventura', 'animacion', 'drama', 'comedia']
peliculas2=peliculas[peliculas.Genero.isin(filtro_generos)]
peliculas2
```

```python
# Filtra el DataFrame por Estudio
#peliculas3 = peliculas2[(peliculas2.Estudio == 'Buena Vista Studios') | (peliculas2.Estudio == 'Fox') | (peliculas2.Estudio == 'Paramount Pictures') | (peliculas2.Estudio == 'Sony') | (peliculas2.Estudio == 'Universal') | (peliculas2.Estudio == 'WB')]
filtro_estudio=['Buena Vista Studios', 'Fox', 'Paramount Pictures', 'Sony', 'Universal', 'WB']
peliculas3=peliculas2[peliculas.Estudio.isin(filtro_estudio)]
peliculas3
```

```python
# Revisa que los filtros se hayan aplicado
print (peliculas2.Genero.unique())
print (peliculas3.Estudio.unique())
print (len(peliculas))
print (len(peliculas2))
print (len(peliculas3))
```

```python
# Define el estilo
sns.set(style="darkgrid", palette="muted", color_codes=True)

# Gráfica los boxplots
ax = sns.boxplot(data=peliculas3, x="Genero", y="Ingresos USA ($mill)", orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)

# Agrega los puntos para mostrar cada observación
sns.stripplot(x="Genero", y="Ingresos USA ($mill)", data=peliculas3, jitter=True, size=6, linewidth=0, hue = 'Estudio', alpha=0.7)

ax.axes.set_title('Ingresos Brutos (%) Domésticos por Género',fontsize=25)
ax.set_xlabel('Genero',fontsize=20)
ax.set_ylabel('Ingresos Bruto(%) USA',fontsize=20)

# Define el lugar de la leyenda
ax.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.show()
```
