Introducción a Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.rcParams["figure.figsize"] = 8,4
import warnings 
warnings.filterwarnings("ignore")
```

```python
datos.head()
```

```python
vis1=sns.distplot(datos["UsuarioDeInternet"], bins = 30)
```

```python
vis2 = sns.boxplot(data=datos, x="GrupoDeIngreso", y="TasaNatalidad")
```

```python
#Ver Ejemplo en Galeria Seaborn
# \ incluimos para salto de linea
```

```python
vis3=sns.lmplot(data=datos, x="UsuarioDeInternet", y="TasaNatalidad", \
                fit_reg=False, hue="GrupoDeIngreso", height=8, aspect=2)
```

```python
vis4=sns.lmplot(data=datos, x="UsuarioDeInternet", y="TasaNatalidad", \
                fit_reg=False, hue="GrupoDeIngreso", height=8, aspect=2,\
               scatter_kws= {"s":100})
```
